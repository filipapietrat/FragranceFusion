```python
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import urllib.request
import ssl
import os

# Disable SSL verification
ssl._create_default_https_context = ssl._create_unverified_context

# Load the DataFrame
df = pd.read_csv('perfumes_batch1.csv')

# Set up WebDriver (assuming Chrome)
driver = webdriver.Chrome()

# Define the directory path
directory_path = "C:\Users\filip\OneDrive\Desktop\IRONHACK\Final_Project\FragranceFusion\App\static"

# Create the directory if it doesn't exist
os.makedirs(directory_path, exist_ok=True)

# Define a function to handle cookies acceptance
def accept_cookies():
    try:
        # Wait for the cookies acceptance element to become visible
        WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//span[contains(text(), "Accept all")]'))
        )
        # Find the cookies acceptance element and click on it
        cookies_button = driver.find_element(By.XPATH, '//span[contains(text(), "Accept all")]')
        cookies_button.click()
    except TimeoutException:
        print("Cookies acceptance element not found or timed out.")

# Define a function to search for and download images
def download_image(row):
    query = f"{row['Brand']} {row['Name']} perfume"
    driver.get(f"https://www.google.com/search?tbm=isch&q={query}")

    # Accept cookies
    accept_cookies()

    # Wait for the image to become visible
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.XPATH, '//div[@id="islrg"]//img'))
    )

    # Find the first image element
    first_image = driver.find_element(By.XPATH, '//div[@id="islrg"]//img')
    image_url = first_image.get_attribute('src')

    # Define a valid file name by replacing invalid characters
    brand_name = str(row['Brand']).replace('/', '_').replace('\\', '_').replace(':', '_').replace('*', '_').replace('?', '_').replace('"', '_').replace('<', '_').replace('>', '_').replace('|', '_')
    perfume_name = str(row['Name']).replace('/', '_').replace('\\', '_').replace(':', '_').replace('*', '_').replace('?', '_').replace('"', '_').replace('<', '_').replace('>', '_').replace('|', '_')
    image_path = f"{directory_path}{brand_name}_{perfume_name}.jpg"

    # Check if the image file already exists
    if os.path.exists(image_path):
        print(f"Image already exists: {image_path}")
        return  # Skip downloading if the image already exists

    # Download the image
    urllib.request.urlretrieve(image_url, image_path)

    # Update the DataFrame with the image path
    df.at[row.name, 'Image_Path'] = image_path

    # Save the DataFrame to CSV every 100 iterations
    if row.name % 100 == 0:
        df.to_csv('perfumes_with_images_partial.csv', index=False)

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    try:
        download_image(row)
    except NoSuchElementException:
        print(f"No image found for {row['Brand']} {row['Name']}. Skipping...")
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        # Save the DataFrame to CSV in case of an error
        df.to_csv('perfumes_with_images_partial.csv', index=False)
        break  # Exit the loop to prevent further iterations

# Add 'image' column with image path to both partial and final DataFrames
df['image'] = df['Image_Path']

# Save the final DataFrame to CSV
df.to_csv('perfumes_with_images.csv', index=False)

# Close the WebDriver
driver.quit()
```
